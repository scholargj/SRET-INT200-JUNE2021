#What is Angular Directive?

Functions that are executed when they are found on the DOM by the Angular Compiler



# Types of Directives in Angular
1. Component Directives
2. Structural Directives
3. Attribute Directives
4. Custom Directives

**What are Structural Directives**

The structural directives alter the layout of the DOM by adding, replacing and removing its elements. 

1. *ngFor 
2. *ngIf
3. ngSwitch

*ngFor: It is a repeater directive that customizes data display. It can be used to display a list of items.

*ngIf: It conditionally includes a template based on the value of an expression coerced to Boolean. When the expression evaluates to true, Angular renders the template provided in a then clause, and when false or null, Angular renders the template provided in an optional else clause.

# Example - 2: Displaying a list of elements  

Create a component **Products** using the command

`ng generate component products `

The code for `products.component.ts` is given below,

````python
import { Component, OnInit } from '@angular/core';

@Component({
  selector: 'app-products',
  templateUrl: './products.component.html',
  styleUrls: ['./products.component.css']
})
export class ProductsComponent implements OnInit
{

  products = [
    {"id":1,"name":"Red Hat Linux","description":"Product of Red Hat Incs.,","Year":"1994","License":"GNU","Version":"7.0"},
    {"id":2,"name":"Android OS","description":"Product of Google","Year":"2008","License":"GNU","Version":"11.10"},
    {"id":3,"name":"Ubuntu OS","description":"UK company","Year":"2018","License":"GNU","Version":"18.04.1"},
    {"id":4,"name":"Windows OS","description":"Microsoft Corporation","Year":"1985","License":"Not FOSS","Version":"10"}
];

  constructor() { }

  ngOnInit(): void {
  }

}
````

The code for `products.component.html` is given below,

````python
<p>Products and their details</p>
<ul>
    <li *ngFor="let product of products; let num = index">
        <h2>#{{num}} {{product.name}} / {{product.Year}}/ {{product.Version}}</h2>
        <p style="text-align: center;font-size:30px;background-color: aqua;color: blueviolet;"> {{product.description}} </p> 
    </li>
    </ul>
````
