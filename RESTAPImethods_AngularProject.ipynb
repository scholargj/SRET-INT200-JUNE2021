The Angular project with default setting has following files:

1. index.html file is the one that is rendering the page.

2. **`<app-root></app-root>`** is called the selector.

3. The app.component.ts contains the selector.

 ```
selector: 'app-root',
```
4. The default template is **`app.component.html`** The view gets reloaded on browser whenever the project is saved.

5. **`package.json`** contains all modules or all dependencies that are currently required to run the project.It contains configuration options for serving, testing and building applications.

6. **`app.module.ts`** contains an entry for all modules, components and services we create in the Angular project.

7. app component is root component and any number of child components can be added here all components reside inside this app folder.


8. **`app.component.ts`** is the file that contain all functionalities written in JS or TS. These functionalities are used by appcomponent.html

9. **`app.component.css`** contains all styling needed for the application. If required to add styles globally to all components then use style.css

# **Step-3. API Url**

**`shared.service.ts`** file

````python
import { Injectable } from '@angular/core';
import {HttpClient} from '@angular/common/http';
import {Observable} from 'rxjs';

@Injectable({
  providedIn: 'root'
})
export class SharedService {

  readonly APIUrl="http://127.0.0.1:8000/";
  readonly PhotoUrl ="http://127.0.0.1:8000/media";
  
  constructor(private http:HttpClient) { }

  
getDepList():Observable<any[]>
{
return this.http.get<any[]>(this.APIUrl + '/department/');
}

addDepartment(val:any)
{
return this.http.post(this.APIUrl + '/department/', val);
}
updateDepartment(val:any)
{
return this.http.put(this.APIUrl + '/department/', val);
}

deleteDepartment(val:any)
{
return this.http.delete(this.APIUrl + '/department/'+val);
}



getEmpList():Observable<any[]>
{
return this.http.get<any[]>(this.APIUrl + '/employee/');
}

addEmployee(val:any)
{
return this.http.post(this.APIUrl + '/employee/', val);
}
updateEmployee(val:any)
{
return this.http.put(this.APIUrl + '/employee/', val);
}

deleteEmployee(val:any)
{
return this.http.delete(this.APIUrl + '/employee/'+val);
}

//Add method to save profile picture

UploadPhoto(val:any){
return this.http.post(this.APIUrl+'/SaveFile',val);
}

//another method to get all department names
getAllDepartmentNames():Observable<any[]>
{
  return this.http.get<any[]>(this.APIUrl+'/department/');
}
}
````
