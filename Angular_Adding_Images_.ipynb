# Step 2: Code the HTML View

The code for the file **`app.component.ts`** is given below.

```python
import { Component } from '@angular/core';

@Component({
  selector: 'app-root',
  templateUrl: './app.component.html',
  styleUrls: ['./app.component.css']
})
export class AppComponent 
{
  title = 'Medical Imaging';
}
```

The code for **`app.component.html`** is given below.

```python
<div class="container">
  <br />
  <div class="row">
      <h1 style="text-align:center;color:darkmagenta"> Welcome to {{ title }}! </h1>
      <br />
  </div>
  <div>
          <h1 style="background-color:aqua;font-size:30px;color:brown"> Computed Tomography : Basic Principles </h1>
          <img src="assets/img/CT1_Anatomy.jpg" alt="Computed Tomography" style="width:600px;height:600px;">
          <br />
          <h1 style="background-color:rgb(21, 255, 0);font-size:30px;color:brown"> MRI Chest Imaging </h1>
          <img src="assets/img/CT2_Anatomy.jpg" alt="Computed Tomography" style="width:800px;height:500px;">
          <br />
          <h1 style="background-color:rgb(255, 230, 0);font-size:30px;color:brown"> CT Chest Anatomy </h1>
          <img src="assets/img/CT3_Anatomy.jpg" alt="Computed Tomography" >
  </div>
  </div>
<router-outlet></router-outlet>
```
