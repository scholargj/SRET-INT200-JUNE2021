# **7. Enable CORS**

**Why CORS need to be enabled?**

By default Django comes with security feature that will block the requests coming  from different domains as it doesnt want everyone to consume the API to access or modify the data.

django-cors-headers is a Django application for handling the server headers required for Cross-Origin Resource Sharing (CORS).

Need to authorize certain domain to consume the API
django-cors-headers is a Django application for handling the server headers required for Cross-Origin Resource Sharing (CORS).


image.png

```python
(myenv) F:\DjangoAngular\DjangoAPIfirst>pip install django-cors-headers
Collecting django-cors-headers
  Downloading django_cors_headers-3.7.0-py3-none-any.whl (12 kB)
Requirement already satisfied: Django>=2.2 in f:\djangoangular\myenv\lib\site-packages (from django-cors-headers) (3.2.4)
Requirement already satisfied: sqlparse>=0.2.2 in f:\djangoangular\myenv\lib\site-packages (from Django>=2.2->django-cors-headers) (0.4.1)
Requirement already satisfied: asgiref<4,>=3.3.2 in f:\djangoangular\myenv\lib\site-packages (from Django>=2.2->django-cors-headers) (3.3.4)
Requirement already satisfied: pytz in f:\djangoangular\myenv\lib\site-packages (from Django>=2.2->django-cors-headers) (2021.1)
Requirement already satisfied: typing-extensions in f:\djangoangular\myenv\lib\site-packages (from asgiref<4,>=3.3.2->Django>=2.2->django-cors-headers) (3.10.0.0)
Installing collected packages: django-cors-headers
Successfully installed django-cors-headers-3.7.0
```

Check **settings.py** and add the following

1. Insert **'corsheaders'** under INSTALLED_APPS

2. To allow all domains to consume the application specify
**`CORS_ORIGIN_ALLOW_ALL = True`**

3. Add **`corsheaders.middleware.CorsMiddleWare`** under MIDDLEWARE


```python
Check settings.py
# Application definition

INSTALLED_APPS = [
    'django.contrib.admin',
    'django.contrib.auth',
    'django.contrib.contenttypes',
    'django.contrib.sessions',
    'django.contrib.messages',
    'django.contrib.staticfiles',
    'corsheaders'
]

CORS_ORIGIN_ALLOW_ALL = True

MIDDLEWARE = [
    'corsheaders.middleware.CorsMiddleWare',
    'django.middleware.security.SecurityMiddleware',
    'django.contrib.sessions.middleware.SessionMiddleware',
    'django.middleware.common.CommonMiddleware',
    'django.middleware.csrf.CsrfViewMiddleware',
    'django.contrib.auth.middleware.AuthenticationMiddleware',
    'django.contrib.messages.middleware.MessageMiddleware',
    'django.middleware.clickjacking.XFrameOptionsMiddleware',
]
```
