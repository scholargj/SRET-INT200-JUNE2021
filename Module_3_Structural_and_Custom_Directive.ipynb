# 1.2 *ngIf, *ngIf then else, *ngIf and *ngIfElse

*ngIf is angular directive used to add or remove an elements from HTML DOM.

Note: Practice the below code in app.component.html.
You need to declare the value or property in app.component.ts

```python
<!--Using a condition bind the value  -->
<h1 *ngIf= 'true'> Hi, Welcome to SRET</h1>
```

```python
<!--How to bind a property using *ngIf directive -->
<h3 *ngIf= 'showme'> Hi, Welcome to Sri Ramachandra Engineering and Technology</h3>
```

```python
<ng-template [ngIf]="showme" [ngIfElse]="info">
  <div>
    <h1 style = color:saddlebrown>You have registered for the seminar ...</h1>
  </div>
</ng-template>

<ng-template #info>
  <div>
   <h2 style=color:springgreen> Hurry Up!!!!!!1, Why not registered? See you Soon......</h2>
  </div>
</ng-template>
```

```python
<div *ngIf="showme; then info; else bye"> This is about Seminar </div>
   <ng-template #info>
  <div>
    <h1 style = color:tomato> Welcome to SRET, You have registered for seminar. Thank you.</h1>
  </div>
</ng-template>

<ng-template #bye>
  <div>
    <h3>Not registered yet? See you soon friend......</h3>
  </div>
</ng-template>
```

#How to customize the style of elements in HTML DOM?

Include the following steps,

1. Include ElementRef in import statement.

2. Pass the reference by parameter. 
Replace the code of constructor as given below,


```python
import { Directive, ElementRef } from '@angular/core';

@Directive({
  selector: '[appCustomStyle]'
})
export class CustomStyleDirective {

  constructor(private el:ElementRef) 
  { 
    el.nativeElement.style.color="blue"
  }

}
```
