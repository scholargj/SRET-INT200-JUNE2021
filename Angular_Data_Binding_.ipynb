```python
<input [(ngModel)]="name" > 
```

**The example for two way data binding**

Edit the file `app.module.ts` as given below,

```python
import { NgModule } from '@angular/core';
import { BrowserModule } from '@angular/platform-browser';

import { AppRoutingModule } from './app-routing.module';
import { AppComponent } from './app.component';

import {FormsModule} from "@angular/forms";
 
@NgModule({
  declarations: [
    AppComponent,
    
  ],
  imports: [   
     BrowserModule, 
     FormsModule,    
     AppRoutingModule
  ],
  providers: [],
  bootstrap: [AppComponent]
})
export class AppModule { }
```

Edit the file `app.component.html` as given below,

```python
<h3 style="text-align:center">This is an example for two way data binding</h3> 
<label> Enter your name:
<input [(ngModel)]='val'/>
</label>
<h4 style="text-align:center">Good Morning, I am {{val}}</h4>
```
