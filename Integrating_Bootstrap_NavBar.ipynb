# Example 2: Integrating bootstrap Navbar in Angular Project

1. Create Angular project **`ng new MyPage`**.
2. Code for **`app.component.html`** is given below.

The class included is `"navbar navbar-expand-lg navbar-light bg-light"`.

**`href="#"`** refer to hyperlink for each item on the navigation bar.

````python
<nav class="navbar navbar-expand-lg navbar-light bg-light">
  <a class="navbar-brand logo" href="#">Medical Imaging Research</a>
  <button class="navbar-toggler" type="button" data-toggle="collapse" data-target="#navbarSupportedContent" aria-controls="navbarSupportedContent" aria-expanded="false" aria-label="Toggle navigation">
    <span class="navbar-toggler-icon"></span>
  </button>

  <div class="collapse navbar-collapse" id="navbarSupportedContent">
    <ul class="navbar-nav mr-auto">
      <li class="nav-item active">
        <a class="nav-link" href="#">Home <span class="sr-only">(current)</span></a>
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#">Data Repository</a>
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#">Applications</a>
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#">Practice</a>
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#">Services</a>
      </li>
      <li class="nav-item">
        <a class="nav-link" href="#">Publications</a>
      </li>
    </ul>
    <ul class="navbar-nav ">
     
      <li class="nav-item dropdown">
        <a class="nav-link dropdown-toggle" data-toggle="dropdown" href="#" role="button" aria-haspopup="true"
        aria-expanded="false">Research Profile</a>
        <div class="dropdown-menu">
          <a class="dropdown-item" href="#">User2</a>
          <a class="dropdown-item" href="#">User3</a>
          
        </div>
      </li>
    </ul>
  </div>
</nav>



<h1 style="font-size:30px;background-color:rgb(30, 255, 199); color:brown"> Hello, Welcome to my page</h1>
<router-outlet></router-outlet>
````

The code for **`app.component.ts`** is given below,

````python
import { Component } from '@angular/core';

@Component({
  selector: 'app-root',
  templateUrl: './app.component.html',
  styleUrls: ['./app.component.css']
})
export class AppComponent {
  title = 'My Page';
  
}
````
