# Reusability:
How to make use of a component without creating again?


The steps involved are :

Make a user component 

Use it inside app component as child 

Pass data from parent to child 

Use for loop and result child component

**Create a new component named user as shown below**





````python
ng generate component user
	 
PS C:\Users\gj\day7> ng generate component user
CREATE src/app/user/user.component.html (19 bytes)
CREATE src/app/user/user.component.spec.ts (612 bytes)
CREATE src/app/user/user.component.ts (267 bytes)     
CREATE src/app/user/user.component.css (0 bytes)      
UPDATE src/app/app.module.ts (637 bytes)
PS C:\Users\gj\day7> 
````

Include the child component inside the `app.component.html `the parent component as shown below

````python
<app-user [hero]="users"></app-user>
````

Define the value for users in `app.component.ts` file

````python
 users ={ 
    name:'peter',
    age:20,  
    location:'Delhi'   ,
  }
````

Include the code given below in ***user.component.ts***

Input is added in import statement to read the values of hero in 

```
@Input() hero: any;
```





````python
mport { Component, OnInit, Input } from '@angular/core';

@Component({
  selector: 'app-user',
  templateUrl: './user.component.html',
  styleUrls: ['./user.component.css']
})
export class UserComponent implements OnInit 
{
@Input() hero: any;
  constructor() 
  {
  
   }

  ngOnInit(): void {
    console.warn(this.hero);
  }

}
````

This line shows the value passed via console 

` console.warn(this.hero);`

use selector tag in `user.component.ts `
**app-user**

Use F12 to see how the console is responding while rendering the page as shown below,

The above figure shows how the child component is used in parent component.

The below code displays the name of hero

````python
<h3>I am child component named "user"</h3>

<p>{{hero.name}}</p>
````

To render the details on the parent component use the code given below,

````python

<!--This is the child component-->

<div class='container'class='box'>
    <h5>{{hero.name}}</h5>
    <h5>{{hero.age}}</h5>
    <h5>{{hero.location}}</h5>
</div>
````

image.png

**Suppose if users are many, then use *ngFor directive to render the data value from child to parent component.**

````python
div *ngFor="let data of users">
  <app-user [hero]="data"></app-user>
</div>
````
