# **Step-3 Code the Show-Department Component**



1. import the sharedService created in step -1

`import { SharedService } from 'src/app/shared.service';`

2. Create a constructor 

```
constructor(private service:SharedService) { }
```

3. Create a empty list to store the data

`DepartmentList:any=[];`


4. Create a method to refresh the Department list. This calls the REST API GET method by passing the URL defined in Django project

```
refreshDepList()
{
  this.service.getDepList().subscribe(data=>{
    this.DepartmentList=data;});
}
```
**subscribe()** method fetches the data to DepartmentList using Asynchronous operation as indicated by symbol =>

5. Call the refreshDepList() in **ngOnInit()** method as this method loads first when component is called.
```
ngOnInit(): void 
  {
    this.refreshDepList();
  }
```

**show-dept.component.ts** file 

````python
import { Component, OnInit } from '@angular/core';
import { SharedService } from 'src/app/shared.service';
@Component({
  selector: 'app-show-dept',
  templateUrl: './show-dept.component.html',
  styleUrls: ['./show-dept.component.css']
})
export class ShowDeptComponent implements OnInit {

  constructor(private service:SharedService) { }
  //create an empty list for department
  DepartmentList:any=[];
  ngOnInit(): void 
  {
    this.refreshDepList();
  }
refreshDepList()
{
  this.service.getDepList().subscribe(data=>{
    this.DepartmentList=data;});
}

}
````

Build the nodeJS and run the development server to serve the application at localhost:4200

# **Step-4 Department page to display the details in tabular form**

1. Create table with headers Department Id and Department Name with options to edit and delete.


2. Edit and Delete buttons are required to Add, update and delete details using REST API POST, PUT and DELETE methods


3. Using `*ngFor` directive populate the table from DepartmentList 

The **show-dept.component.html** file is given below,

````python
<h1> This is show department page </h1>

<table class="table table-striped"> 
    <thead>
        <tr>
            <th>Department Id</th>
            <th>Department Name</th>
            <th>Options</th>
        </tr>
    </thead>
    <tbody>
        <tr *ngFor="let dataItem of DepartmentList">
            <td>{{dataItem.DepartmentId}}</td>
            <td>{{dataItem.DepartmentName}}</td>
            <td>
                <button type="button" class="btn btn-light mr-1">
                    Edit
                </button>
                <button type="button" class="btn btn-light mr-1">
                    Delete
                </button>
            </td>
        </tr>
    </tbody>
</table>
````
