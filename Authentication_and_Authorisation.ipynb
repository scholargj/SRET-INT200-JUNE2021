# Compare Authentication and Authorisation
**Authentication**

Basic:	Checks the person's identity to grant access to the system.

Include process of : Verifying user credentials

Order of the process: Authentication is performed at the very first step.

**Authorization**

Basic:	Checks the person's privileges or permissions to access the resource. 

Include process of :Validating the user permission

Order of the process: Authorisation is usually performed after Authentication.


*Example: Authentication*
In online banking applications, the identity of the person is first determined with the help of the user ID and password.

*Example: Authorisation*
In a multi-user system, the administrator decides what preivileges or access rights do each other have.
