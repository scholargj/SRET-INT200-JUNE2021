# **Step - 2 Add Routing**

**Add the import statements**
```
import {EmployeeComponent} from './employee/employee.component';
import {DepartmentComponent} from './department/department.component';
```

**Add the path to display the pages in Routes array**

```
const routes: Routes = [
{path:'employee',component:EmployeeComponent},
{path:'department', component:DepartmentComponent}
```



];

**app-routing.module.ts** file

````python
import { NgModule } from '@angular/core';
import { RouterModule, Routes } from '@angular/router';

import {EmployeeComponent} from './employee/employee.component';
import {DepartmentComponent} from './department/department.component';

const routes: Routes = [
{path:'employee',component:EmployeeComponent},
{path:'department', component:DepartmentComponent}

];

@NgModule({
  imports: [RouterModule.forRoot(routes)],
  exports: [RouterModule]
})
export class AppRoutingModule { }
````

# **Step-4 Add Navigation bar**
Navigations bar will have controls to navigate to department and Employee pages

**`app.component.html`** file

````python
<div class = "container">

<h3 class="d-flex justify-content-center">Angular-Django REST API</h3>

<h5 class="d-flex justify-content-center">Employee Management Portal</h5>

<nav class ="navbar navbar-expand-sm bg-light navbar-dark">
<ul class="navbar-nav">
    <li class="nav-item">
        <button routerLink="department"
        class="m-1 btn btn-light btn-outline-success" Button>Departments</button>
    </li>

    <li class="nav-item">
        <button routerLink="employee"
        class="m-1 btn btn-light btn-outline-success" Button>Employees</button>
    </li>
</ul>
</nav>
<router-outlet></router-outlet>
</div>
````
