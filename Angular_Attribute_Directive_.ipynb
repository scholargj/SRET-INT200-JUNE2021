# ngClass and ngStyle directive
`NgClass & NgStyle` are Angular Directives that allow us to conditionally apply one-to-many classes/styles to an element. 

Native style and class attributes apply one-to-many classes/styles statically. 

Angular's class and style bindings only allow a single class or style to be conditionally applied at a time.

**Note:** Use the code in `app.component.html`


```python
<div class="content">
<button [ngStyle]="{'font-size.px':20,'font-style':'Italic', 'color':'red'}">Style option1</button>
</div>

<div class="container">
<button [ngStyle]="AddStyle()">Style option 2</button>

</div>
<div class ="container">
<button [ngClass]="['btn-primary','btn-outline-warning','btn-red']">ngClass Array</button>
<button [ngClass]="{'btn-primary':false,'btn-outline-primary':true,'btn-red':true}">ngClass Object</button>
<button [ngClass]="AddClass()">ngClass Function</button>
</div>
```

**Note:** Use the code in `app.component.ts`

```python
 AddStyle()
  {
  return { 'font-size.px':20,'font-style':'italic', 'color':'red'  }

  }

AddClass()
{
  return{'btn-warning':true,'btn-italic':true,'btn-green':true}
}
```

#Example: How to conditionally style the text?

Consider the names of actors and their character names with some additional details is to displayed with a conditional background.

Use the below code in `app.component.ts`

```python
Character= [
  {
     actor_name: 'Peter Dinklage',
     character_name: 'Tyrion Lannister',
     gender: 'Male',
     status: 'Alive'
  },
  {
     actor_name: 'Sean Bean',
     character_name: 'Ned Stark',
     gender: 'Male',
     status: 'Dead'
  },
  {
     actor_name: 'Emilia Clark',
     character_name: 'Khaleesi',
     gender: 'Female',
     status: 'Alive'
  },
  {
     actor_name: 'Catelyn Stark',
     character_name: 'Michelle Fairley',
     gender: 'Female',
     status: 'Dead'
  }
];
```

Use the below code in `app.component.html`

```python
<!-- ngStyle Demonstration-->
<div class="container">
  <h4>NgStyle</h4>
  <div *ngFor="let celeb of Character">
     <p [ngStyle]="{'background-color':celeb.status === 'Dead' ? 'red' : 'green' }">
    {{ celeb.actor_name }} ({{ celeb.character_name }}) is {{celeb.gender}} character and {{celeb.status}}
     </p>
   </div>
</div>
```
